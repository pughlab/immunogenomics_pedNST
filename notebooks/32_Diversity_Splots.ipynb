# Main

```r
load(file = paste0(datapath, "diversity/metadata_IC_TRB.RData"))
```

```r
Div_df.cohortsB <- median.cohorts.fx(metadata_IC_TRB)
sorted_df_listB <- sort.cohorts.fx(metadata_IC_TRB, Div_df.cohortsB)
TRBplot <- Splot.fx(sorted_df_listB, "TRb estimated Shannon diversity")
```

```r
TRBplot + annotation_logticks(sides = "l") +
scale_y_continuous(trans = "log10", 
                   limits = c(1, 10000),
                   breaks = c(1,10,100,1000,10000),
                  labels = scales::label_number(accuracy = 1)) +
ggtitle(~underline("TRb estimated Shannon diversity"))
```

```r
pdf(file = paste0(plotpath, "Splot_TRB.pdf"),
    width = 8, 
    height = 8,
    useDingbats = FALSE)
TRBplot + annotation_logticks(sides = "l") +
scale_y_continuous(trans = "log10", 
                   limits = c(1, 10000),
                   breaks = c(1,10,100,1000,10000),
                  labels = scales::label_number(accuracy = 1)) +
ggtitle(~underline("TRb estimated Shannon diversity"))
dev.off()
```
