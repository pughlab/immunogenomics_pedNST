# TRB

```r
filelist <- list.files(mountpath, pattern = "CLONES_TRB", recursive = T)
```

```r
filelist <- filelist[!grepl("TCGA", filelist)]
```

```r
trb_list <- immunelistfx(filelist, mountpath, "TRB")
```

```r
length(trb_list)
```

```r
trb_list <- trb_list[sapply(trb_list, function(x) length(unlist(x))) > 1] # remove files with one clonotype
                            
length(trb_list)
```

```r
trb_list <- trb_list[sapply(trb_list, function(x) var(unlist(x))) > 0] # remove files no variance                        
```

```r
length(trb_list)
```

```r
save(trb_list, file = paste0(datapath, "diversity/trb_list_ped.RData"))
```

```r
load(file = paste0(datapath, "diversity/trb_list_ped.RData"))
```

```r
div_trb <- Divstats.fx(trb_list, "TRB")
```

```r
save(div_trb, file = paste0(datapath, "diversity/div_trb_ped.RData"))
```

```r
head(div_trb)
```

```r
div_trb <- as.data.frame(div_trb)
```

```r
div_trb$filename <- rownames(div_trb)

div_trb$sample_id <- gsub("CLONES_TRB_", "", div_trb$filename)
div_trb$sample_id <- gsub(".txt", "", div_trb$sample_id)
div_trb$sample_id <- gsub(".*/", "", div_trb$sample_id)

div_trb$sample_id[ grepl("7316", div_trb$sample_id)] <- gsub("_.*", "", 
                                                             div_trb$sample_id[grepl("7316", div_trb$sample_id)])
```

```r
div_trb$sample_id
```

```r
metadata_IC
```

```r
metadata_IC_TRB <- merge(metadata_IC, div_trb, by = "sample_id")
```

```r
table(metadata_IC_TRB$group)
```

```r
dim(metadata_IC_TRB)
```

```r
save(metadata_IC_TRB,
         file = paste0(datapath, "diversity/metadata_IC_TRB.RData"))
```
